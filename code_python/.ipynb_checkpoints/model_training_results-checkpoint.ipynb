### Stacked

```python
lr = LogisticRegression(max_iter=1000)
lr.set_params(**bs_logreg.best_params_)
rf = RandomForestClassifier(random_state=1, n_jobs=-1)
rf.set_params(**bs_rf.best_params_)
xgb = XGBClassifier(nthread=-1, seed=1)
xgb.set_params(**bs_xgb.best_params_)
nn = MLPClassifier(activation='logistic',
    learning_rate='adaptive', random_state=1)
nn.set_params(**gs_nn.best_params_)

level0_models = [lr, rf, xgb, nn]

S_train, S_test = stacking(level0_models, X_train[:,rfe.support_], y_train, 
    X_test[:,rfe.support_], regression=False, mode='oof_pred_bag',
    needs_proba=True, save_dir=None, metric=roc_auc_score, n_folds=5,
    stratified=True, shuffle=True, random_state=0, verbose=0)

level1_model = LogisticRegression()
level1_model = level1_model.fit(S_train, y_train)
```

```python
y_pred = level1_model.predict(S_test)
y_probas_ = level1_model.predict_proba(S_test)
print('Confusion Matrix:\n',confusion_matrix(y_test, y_pred))
print('Classification Report:\n',classification_report(y_test, y_pred))
print('Area Under ROC:',roc_auc_score(y_test, y_probas_[:, 1]))
fpr, tpr, _ = roc_curve(y_test, y_probas_[:, 1])
plt.plot(fpr, tpr, marker='.', label='Stacked')
plt.plot(np.linspace(0,1,101), np.linspace(0,1,101), linestyle='--', label='Random')
plt.legend()
plt.show()
```
